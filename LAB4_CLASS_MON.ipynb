```python
# 널값 확인
loan.isnull() # 널값이 있으면 True
loan.isna() # 널값이 있으면 True
loan.notnull() # 널값이 없으면 True
```

```python
# 널값 시각화
sns.heatmap(loan.isnull(), cbar=False)
plt.show()
```

```python
# 널값이 아닌 데이터셋 생성
loan_notnull = loan[loan.notnull()]
loan_notnull.shape
```

```python
# 널값 요약
print('널값 요약')
print(loan.isnull().sum())
print('\n전체 널값의 갯수: ', loan.isna().sum().sum())
```

```python
# 널값 제거
loan.dropna()
```

```python
# 널값 0으로 채우기
loan.fillna(0)
```

```python
# gender 널값 처리
loan['Gender'].mode() # 최빈값 리턴
loan.Gender.value_counts() # 빈도표
loan.Gender.fillna(loan['Gender'].mode()[0]) # Male로 채우기
loan.Gender.fillna('Male') # Male로 채우기
```

```python
loan.columns
```

```python
# 범주형 변수의 널값 한꺼번에 채우기
col_lst = ['Gender','Married','Dependents','Education','Self_Employed','Loan_Amount_Term','Credit_History','Property_Area']
for col in col_lst:
  loan[col].fillna(loan[col].mode()[0], inplace=True)
```

```python
# 처리한 후 널값 확인
loan.isna().sum()
```

```python
# LoanAmount 히스트그램으로 패턴파악
loan.LoanAmount.hist()
plt.show()
```

```python
# LoanAmount 널값 처리
loan.LoanAmount.fillna(loan.LoanAmount.median, inplace=True)
```

```python
# 널값 확인
loan.isna().sum()
```

# 인코딩

```python
import sys
sys.setrecursionlimit(10000)
```

```python
# Property_Area: 원핫인코더, get_dummies 이용
dummy = pd.get_dummies(loan.Property_Area)
dummy.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
dummy = ohe.fit_transform(loan['Property_Area'].values.reshape(-1,1)).toarray()
dummy = pd.DataFrame(dummy, columns=ohe.categories_)
dummy.head()
```

```python
# Gender, Married, Education, Self_Employed: 레이블 인코더 이용 (for loop)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
loan.Gender = le.fit_transform(loan.Gender)
```

```python
loan.Married = le.fit_transform(loan.Married)
```

```python
loan.Education = le.fit_transform(loan.Education)
```

```python
loan.Self_Employed = le.fit_transform(loan.Self_Employed)
```

```python
loan.head()
```

```python
# Gender, Married, Education, Self_Employed: 레이블 인코더 이용 (for loop)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
col_lst = ['Gender','Married','Education','Self_Employed']
for col in col_lst:
  loan[col] = le.fit_transform(loan[col])
loan.head()
```

```python
# Loan_Status ({'Y':1, 'N':0}, inplace=True), Dependents ('3+','3',inplace=True): replace 이용해서 인코딩
loan.Loan_Status.replace({'Y':1, 'N':0}, inplace=True)
```

```python
loan.Dependents.replace({'3+':3}, inplace=True)
```

```python
# 인코딩한 후 데이터 타입을 int32로 변환 (astype)
loan.dtypes
loan.Dependents = loan.Dependents.astype('int32')
```
