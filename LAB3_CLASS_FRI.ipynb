# 2019년 중구 월별 교통사고 발생건수

```python
# 필요한 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 파일 열기
traffic = pd.read_table('/content/report1.txt')
traffic.head()
```

```python
# 데이터 자르기
junggu = traffic[traffic.자치구별=='중구'][traffic.구분=='발생건수'].iloc[:, -12:]

# 데이터 타입 변환
junggu = junggu.astype('int')
```

```python
# 한글처리 코드
# !apt -y install fonts-nanum > /dev/null
# !fc-cache -fv > /dev/null
# !rm -rf ~/.cache/matplotlib > /dev/null
plt.rc('font', family='NanumBarunGothic')
```

```python
# 1차원으로 변환해서 그래프 그리기
plt.plot(junggu.values[0], marker='<', color='r')
plt.title('2019년 중구 월별 교통사고 발생건수')
plt.xlabel('월')
plt.ylabel('발생건수')
plt.grid()
plt.ylim(0,120)
plt.xticks(range(12), ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'])
plt.show()
```

# 2009년~2019년 서울시 교통사고 발생건수와 부상자수 산점도

```python
traffic_all = pd.read_table('/content/report2.txt')
traffic_all.head()
```

```python
# 원하는 부분 자르기
occurrence = traffic_all[traffic_all.구분=='발생건수'][traffic_all.자치구별=='합계'].loc[:,'1월':'12월']
injured =  traffic_all[traffic_all.구분=='부상자수'][traffic_all.자치구별=='합계'].loc[:,'1월':'12월']
```

```python
occurrence
```

```python
injured
```

```python
# 중간 쉼표 제거
occurrence = occurrence.apply(lambda x: x.str.replace(',', ''))
injured = injured.apply(lambda x: x.str.replace(',', ''))

# 문자열을 숫자로 변환
occurrence = occurrence.astype('int')
injured = injured.astype('int')
```

```python
# 일렬로 만들기
occurrence = occurrence.values.flatten()
injured = injured.values.flatten()
```

```python
# 산점도 그리기
plt.scatter(occurrence, injured)
plt.xlabel('발생건수')
plt.ylabel('부상자수')
plt.grid()
plt.xlim(0, 5000)
plt.ylim(0, 6000)
plt.title('2009년에서 2019년 서울시 발생건수와 부상자수 산점도')
plt.show()
```
