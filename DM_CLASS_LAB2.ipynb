# Chipotle 예제
# 제출: 28번, 29번

```python
# 데이터셋을 임포트해서 chipo라고 하시오.
chipo = pd.read_excel('/content/chipotle.xlsx')

# 처음 10줄
chipo.head(10)

# 관측치의 갯수는?
chipo.shape[0]

# 컬럼의 수는?
chipo.shape[1]

# 모든 컬럼의 이름을 출력
chipo.columns

# 데이터셋은 어떻게 인덱스 되었나요?
chipo.index

# 전체 주문 아이템수는? Quantity의 합
chipo.quantity.sum()

# 데이터셋 안에 있는 기간에 대한 수익은 얼마인가요?
(chipo.item_price * chipo.quantity).sum()
```

```python
# 가장 많이 주문된 아이템은? Item_name에 따라 groupby된 것중 가장 카운트가 많이 된 것을 찾음
chipo.item_name.value_counts().index[0]

# 가장 많이 주문된 아이템은 몇개나 주문됐나요? Groupby된것의 카운트수
chipo.item_name.value_counts()[0]

# choice_description 컬럼에서 가장 많이 주문된 아이템은?
chipo.choice_description.value_counts().index[0]

# choice_description에 따라 groupby된 것중 가장 카운트가 많은 것
chipo.choice_description.value_counts()[0]

# 얼마나 많은 다른 종류의 아이템이 팔렸나요?
chipo.item_name.nunique()
```

# Occupation 예제

```python
# 데이터셋을 임포트해서 users라는 변수에 저장한후 user_id를 인덱스로 사용하시오.
users = pd.read_csv('/content/occupation.txt', sep='|')
users

# 처음 25줄
users.head(25)

# 마지막 10줄
users.tail(10)

# 관측치의 수
users.shape[0]

# 컬럼의 수
users.shape[1]
```

```python
# 모든 컬럼의 이름
users.columns

# 데이터셋은 어떻게 인덱스 되었나요?
users.index

# 각 컬럼의 데이터타입은?
users.dtypes

# 직업컬럼만 출력
users.occupation
```

```python
# 기술통계를 이용하여 데이터요약
users.describe()

# 모든 컬럼을 요약
users.describe(include='all')

# 직업 컬럼만 요약
users.occupation.describe()

# 모든 사용자의 나이의 평균
users.age.mean()
```

```python
# 직업의 종류는?
users.occupation.nunique()

# 가장 많은 직업은?
users.occupation.value_counts().head(1)

# 가장 적게 나오는 나이는?
users.age.value_counts().tail()
```
